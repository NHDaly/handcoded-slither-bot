# Random Slither.io

```python
import universe, gym
import numpy as np
```

### Load the environment

```python
env = gym.make('internet.SlitherIO-v0')
```

```python
# The SlitherIO environment doesn't contain useful sampling of its action
# space. Probably because it's still new.
game_screen_xrange = (18,522)
game_screen_yrange = (86,391)
def sample_game_screen_pos():
    '''Sample an (x,y) point from the playable game window.'''
    return (np.random.randint(*game_screen_xrange),
            np.random.randint(*game_screen_yrange))
def sample_action():
    '''Sample an action ready to send to the environment.

    All Slither actions are mouse actions, either a mouse point or a
    primary-button mouse click.
    '''
    point = sample_game_screen_pos()
    clicking = np.random.randint(0,2) # Yes or no clicking.
    return universe.spaces.vnc_action_space.vnc_event.PointerEvent(x=point[0],
                                                                   y=point[1],
                                                                   buttonmask=clicking)
print(sample_action())
```

```python
s = env.reset()
```

```python
while True:
    a = sample_action()
    s1,r,d,_ = env.step([[a]])
    print(s1, r, d)
    import time
    time.sleep(0.5) # Send an action every .5 seconds
```

# Output
This gets you a random Slither bot, which lives surprisingly long! It's a pretty sparse field, it turns out...
